```python
def plot_by_feature (data, year=None, agg='sum'):
    dd=data
    if year: dd = dd[dd.datetime.dt.year == year]
    dd['hour']=dd.datetime.dt.hour
        
    group_by = dd.groupby(['hour', 'workingday'])['count'].agg(agg).unstack()
    return group_by.plot(kind='bar', figsize=(10,7))
        
        
```

```python
plot_by_feature(train_bike, year=2011)
```

```python
plot_by_feature(train_bike, year =2012)
```

```python
def plot_by_year (attribute, data, title, year=None):
    dd = data
    if year: dd[dd.datetime.dt.year == year]
    
    dd['hour'] = dd.datetime.dt.hour
    dd['month'] = dd.datetime.dt.month
    dd['year'] = dd.datetime.dt.year
    
    group_by = dd.groupby([attribute, 'year'])['count'].agg('sum').unstack()
    return group_by.plot(kind='bar', figsize=(10,7), title=title)
    
    
```

```python
plot_by_year('month', train_bike, 'Rent bikes per month in 2011 and 2012', 2011)
```

```python
plot_by_year('workingday', train_bike, 'rent per bike by year')
```

```python
plot_by_year('hour', train_bike, 'rent of bikes per hour')
```

```python
def plot_by_hour (data, year=None):
    dd = data
    if year: dd = dd[dd.datetime.dt.year==year]
    dd['hour'] = dd.datetime.dt.hour
    
    hours = {}
    for hour in range (24):
        hours[hour] = dd[dd.hour == hour] ['count'].values
    plt.figure(figsize=(17,7))
    plt.boxplot([hours[hour] for hour in range (24)])
    
```

```python
plot_by_hour(train_bike, 2011)
```

```python
plot_by_hour(train_bike, year=2012)
```

```python
def numeric_to_categorical(data):
    if 0<= data <6:
        return 0
    elif 6<= data <12:
        return 1
    elif 12 <= data <18:
        return 2
    elif 18<= data <24:
        return 3
```

```python
train_bike['hour'] = train_bike['hour'].apply(numeric_to_categorical)
train_bike.head()
```

```python
def select_features (data):
    black_list = ['season', 'holiday', 'registered', 'count']
    return [feature for feature in data.columns if feature not in black_list]
```
